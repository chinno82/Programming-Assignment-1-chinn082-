Cropping and Resize Images 

```python
#Cropping and Resize Images 

def get_bounding_boxes(annot):
  xml=annot
  tree=e.parse(xml)
  root=tree.getroot()
  objects=root.findall('object')
  bbox=[]
  for o in objects:
    bndbox=o.find('bndbox')
    xmin=int(bndbox.find('xmin').text)
    ymin=int(bndbox.find('ymin').text)
    xmax=int(bndbox.find('xmax').text)
    ymax=int(bndbox.find('ymax').text)
    bbox.append((xmin,ymin,xmax,ymax))
  return bbox


def cropping(img_path,ant_path,crp_path):
  img=cv2.imread(img_path)
  c=get_bounding_boxes(ant_path)
  blim=c[0]
  final_data=img[blim[1]:blim[3],blim[0]:blim[2]]   
  final_data=cv2.resize(final_data,dsize=(331 ,331)) 
  final_img=cv2.cvtColor(final_data,cv2.COLOR_RGB2BGR)
  cv2.imwrite(crp_path,final_img)
  #final_img=cv2.imsave(final_data)  
  #final_data.save(crp_path)
```

```python
#canvas=plt.figure(figsize=(10,25))

img1_c1=cv2.imread(r'D:\DM Assignments\affenpinscher\n02110627_1581.jpg')
img2_c1=cv2.imread(r'D:\DM Assignments\affenpinscher\n02110627_2416.jpg')
img3_c2=cv2.imread(r'D:\DM Assignments\cairn\n02096177_91.jpg')

img1_g1 = cv2.cvtColor(img1_c1,cv2.COLOR_BGR2GRAY)
img2_g1 = cv2.cvtColor(img2_c1,cv2.COLOR_BGR2GRAY)
img3_g2 = cv2.cvtColor(img3_c2,cv2.COLOR_BGR2GRAY)



eq1=cv2.equalizeHist(img1_g1)
eq2=cv2.equalizeHist(img2_g1)
eq3=cv2.equalizeHist(img3_g2)


histo1 = cv2.calcHist([eq1], [0], None, [256], [0, 256])
histo2 = cv2.calcHist([eq2], [0], None, [256], [0, 256])
histo3 = cv2.calcHist([eq3], [0], None, [256], [0, 256])

a1= []
for i in histo1:
    a1.append(i[0])

a2= []
for i in histo2:
    a2.append(i[0])

a3= []
for i in histo3:
    a3.append(i[0])


#canvas.add_subplot(4,3,1)
#plt.imshow(img1_c1)

#canvas.add_subplot(4,3,2)
#plt.imshow(img2_c1)

#canvas.add_subplot(4,3,3)
#plt.imshow(img3_c2)
```

```python
# Euclidean Distance
from scipy.spatial import  distance as dist
print('Euclidean Distance for the same classes: ',dist.euclidean(a1, a2))
print('Euclidean Distance for the different classes: ',dist.euclidean(a1, a3))
```

```python
#Manhattan Distance
from scipy.spatial.distance import cityblock
print('Manhattan Distance for the same classes: ',cityblock(a1,a3))
print('Manhattan Distance for the same classes: ',cityblock(a2,a3))
```

```python
# Battacharya Distance
print('Battacharya Distance for the same classes: ',cv2.compareHist(histo1,histo2,cv2.HISTCMP_BHATTACHARYYA))
print('Battacharya Distance for the different classes: ',cv2.compareHist(histo1,histo3,cv2.HISTCMP_BHATTACHARYYA))
```

```python
# Histogram Intersection
print('Histogram Intersection for the same classes: ',cv2.compareHist(histo1,histo2,cv2.HISTCMP_INTERSECT))
print('Histogram Intersection for the different classes: ',cv2.compareHist(histo1,histo3,cv2.HISTCMP_INTERSECT))
```

Image Feature Descriptor: ORB (Oriented FAST and Rotated BRIEF) 

```python
# (e) Image Feature Descriptor: ORB (Oriented FAST and Rotated BRIEF) 
edge_td = 20
pt_size = 16
kp = 40

img = cv2.imread(r'D:\DM Assignments\Cardigan\n02113186_9809.jpg' )
orb = cv2.ORB_create(edgeThreshold= edge_td,patchSize=pt_size, nlevels=2, fastThreshold=30,scaleFactor=2, WTA_K=4,scoreType=cv2.ORB_HARRIS_SCORE,firstLevel=0, nfeatures=kp)

kp = orb.detect(img,None)

kp, des = orb.compute(img, kp)

img2 = cv2.drawKeypoints(img, kp, None, color=(0,0,255), flags=0)
plt.imshow(img2), plt.show()

```

Dimensionality reduction (using Principal Component Analysis, PCA)

```python
import os
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

cls_1 = r"D:\DM Assignments\Cardigan"
cls_2 = r"D:\DM Assignments\cairn"

#mages to grayscale pixel intensity histograms 
def img_hist(img_path):
    img= cv2.imread(img_path)
    gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    hist= cv2.calcHist([gray], [0], None, [256], [0, 256])
    return hist.flatten()


def folder_hist(f1, f2):

    f1_images = [os.path.join(f1, filename) for filename in os.listdir(f1)]
    f2_images = [os.path.join(f2, filename) for filename in os.listdir(f2)]
    hists = [img_hist(file) for file in f1_images + f2_images]
    return hists, len(f1_images), len(f2_images)

def pca_hist(hists):
    norm_hist = StandardScaler().fit_transform(hists)
    pca_transfer= PCA(n_components=2)
    return pca_transfer.fit_transform(norm_hist)

# performing Principal Component Analysis (PCA) dimensionality reduction 
def dim_hist(principal_components, c1, c2):
   
    plt.figure(figsize=(12, 6))
    plt.scatter(principal_components[:c1, 0], principal_components[:c1, 1], c='g')
    plt.scatter(principal_components[c1:, 0], principal_components[c1:, 1], c='b')
   
    plt.show()

# visualization using pca
def pca_visual(f1, f2):
    hists, c1, c2 =folder_hist(f1, f2)
    principal_components = pca_hist(hists)
    dim_hist(principal_components, c1, c2)




pca_visual(cls_1, cls_2)
```

In the figure there are two different classes refered with two different colours. In the mid region, there are so many points intersected 
 And also there are maximum points with intersected . Thus we can conclude that are not seperable.
